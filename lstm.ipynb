```python
import os
os.environ["CUDA_VISIBLE_DEVICES"]="0"
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2' # 디버그 메시지 끄기
import tensorflow as tf
# gpu 사용 확인
print(tf.test.gpu_device_name())
from tensorflow import keras
from keras.optimizers import Adam
from tensorflow.keras import layers
from tensorflow.keras.models import Model
from tensorflow.keras.layers import LSTM, Input, Dense, Embedding, Dropout
import matplotlib.pyplot as plt # 데이터 시각화

import natsort
import json

from angle_out import out
import numpy as np

# tf.keras.backend.clear_session()

def lstm():
    model = keras.Sequential()
    model.add(layers.LSTM(40, activation='relu', input_shape=(4, 10), return_sequences=False))
    model.add(layers.Dropout(0.2))
    model.add(layers.Dense(3, activation='softmax'))
    
    model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
    
    return model
```
